```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [12, 18, 22, 28 ,35]
}

df = pd.DataFrame(data)


x = df['X']
y = df['Y']
x2=df['X']**2
xy=df['X']*df['Y']
n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)
x2_mean=np.mean(x2)
xy_mean=np.mean(xy)

numerator = xy_mean-(x_mean*y_mean)
denominator = x2_mean-(x_mean**2)

b1 = numerator / denominator
b0 = y_mean - b1 * x_mean

print("b1:", b1)
print("b0:", b0)

y5=b0+b1*5
print("Predicted value for x=5:",y5)
y7=b0+b1*7
print("Predicted value for x=7:",y7)
```

```python
plt.scatter(x, y, label='Data Points')
plt.plot(x, b0 + b1 * x, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regression Plot')
plt.legend()
plt.show()
```

Linear Regression In Matrix Form


```python
X = np.column_stack((np.ones(len(x)), x))

beta = np.linalg.inv(X.T @ X) @ X.T @ y

b0 = beta[0]
b1 = beta[1]

print("Slope (b1):", b1)
print("Intercept (b0):", b0)

x_new = 5
y_predicted = b0 + (b1 * x_new)

print("Predicted value for x = 5:", y_predicted)
```
